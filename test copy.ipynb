```python
import csv
# 创建文件
f = open('cpu自营2.csv',mode='w',encoding='utf-8',newline='')
#字典写入方法
csv_writer = csv.DictWriter(f, fieldnames=[
        '商品名称',
        '价格',
        '店铺名称',
        '颜色',
        '评论数',
        '图片',
        '店铺ID',
        '卖点',
        'skuId',
        '总销量',
        '商品ID'
])
```

```python
import re

#访问网站
dp.get('https://search.jd.com/Search?keyword=&enc=utf-8')
#输入搜索内容
dp.ele('css:.jd_pc_search_bar_react_search_input').input('cpu自营')
# 监听数据
dp.listen.start('api?appid=search-pc-java&t')
dp.ele('css:.jd_pc_search_bar_react_search_btn').click()
# 下滑页面到底部
dp.scroll.to_bottom()
#等待数据包加载
resp_ = dp.listen.wait(10)

for resp in resp_:
    #获取返回数据内容
    json_data = resp.response.body
    # 判断商品数据接口
    keys = json_data.keys()
    if 'abBuriedTagMap' in keys:
        # 提取商品数据
        #print(json_data)
        #字典取值:提取商品信息所在的列表
        wareList = json_data['data']['wareList']
        #for循环遍历，提取列表里面的元素
        for index in wareList:
            # 处理商品名称
            title = index['wareName']
            new_title = re.sub('<.*?>','',title)
            #在循环中提取具体商品数据信息
            dit={
                '商品名称':new_title,
                '价格':index['jdPrice'],
                '店铺名称':index['shopName'],
                '颜色':index['color'],
                '评论数':index['commentFuzzy'],
                '图片':index['imageurl'],
                '店铺ID':index['shopId'],
                '卖点':index['sellingPoint'],
                'skuId':index['skuId'],
                '总销量':index['totalSales'],
                #'店铺链接':index['shopLink'],
                '商品ID':index['wareId'],
                #'商品链接':index['wareLink'],
            }
            # 写入数据
            csv_writer.writerow(dit)
            print(dit)
        
```
